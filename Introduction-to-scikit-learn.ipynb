## 0. An end-to-end Scikit_learn workflow

```python
#1. Get the data ready
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease
```

```python
# Create X (features matrix)
X = heart_disease.drop('target',axis=1)

# Create y (labels)
Y = heart_disease["target"]
```

```python
import warnings
warnings.filterwarnings("default") #optional
```

```python
# 2. Choose the right model and hyperparameteres
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100) #clf = classifier

#we'll keep the default hyperparameters
#clf.get_params()
```

```python
# 3. Fit the model to the training data
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2) #if 1000 datas are there 200 will be used for testing and rest 800 will be used for training 
```

```python
import sklearn
sklearn.show_versions()
```

```python
clf.fit(X_train,Y_train); #Have to fit our training variables to the model
```

```python
X_train
```

```python
X_test
```

```python
# Make the prediction
Y_preds = clf.predict(X_test)
Y_preds
```

```python
Y_test
```

```python
# 4. Evaluate the model on the training data and test data
clf.score(X_train,Y_train) #return mean accruacy
```

```python
clf.score(X_test,Y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print(classification_report(Y_test,Y_preds))
```

```python
confusion_matrix(Y_test,Y_preds)
```

```python
accuracy_score(Y_test,Y_preds)
```

```python
# 5. Improve a model
# Try different amount of n_estimators
np.random.seed(42)
for i in range (10,100,10):
    print(f"Trying model with {i} estimators...")
    clf = RandomForestClassifier(n_estimators=i).fit(X_train,Y_train) #training is happening here 
    print(f"Model accuracy on test set: {clf.score(X_test,Y_test) * 100:.2f}%")
    print(" ")
    
#clf is a model and we train thinks there
```

```python
#6. Save a model and load it 
import pickle
pickle.dump(clf,open("random_forest_model_1.pkl","wb"))
```

```python
loaded_model = pickle.load(open("random_forest_model_1.pkl","rb"))
```

```python
loaded_model.score(X_test,Y_test)
```
